Есть два набора точек - *x_hw* и *y_hw*. В рамках домашней работы нужно
* визуализировать набор точек
* найти коэффициенты регрессии $w_0, w_1$ по шагам, как в уроке
* посчитать предсказание в виде $\hat{y} = w_0 + w_1x$ и визуализировать его вместе с точками *x_hw* и *y_hw*

```python
import pandas as pd
```

```python
def ndprint(a, format_string ='{0:.2f}'):
    """Cписок в красивом виде"""
    return [format_string.format(v,i) for i,v in enumerate(a)]
```

```python
x_hw = [50, 60, 70, 100]
y_hw = [10, 15, 40, 45]
```

```python
# Create visualization on scatter plot
df = pd.DataFrame({
    'x_hw': x_hw,
    'y_hw': y_hw
})


# Create a scatter plot
df.plot.scatter(x='x_hw', y='y_hw', title='Scatter Plot of x and y')
```

```python
from numpy.linalg import inv
import numpy as np

x_hw = np.array(x_hw)
y_hw = np.array(y_hw)

X = np.c_[np.ones(x_hw.shape[0]), x_hw]

w_analytic = inv(X.T.dot(X)).dot(X.T).dot(y_hw)

print("Аналитически определённые коэффициенты: \n%s" % ndprint(w_analytic))
```

```python
print("y = {:.2f}x + {:.2f}".format(w_analytic[0], w_analytic[1]))
```

```python
# Create a scatter plot with x_hw, y_hw and the regression line
import matplotlib.pyplot as plt
plt.scatter(x_hw, y_hw)
plt.plot(x_hw, X.dot(w_analytic), color='red')
```
